```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('/content/zomato.csv', encoding = "ISO-8859-1")
df.head()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
new_colum =pd.read_excel('/content/Country-Code.xlsx')
new_colum.head()
```

```python
zomato = pd.merge(df,new_colum,on='Country Code', how="left", )
```

```python
zomato.columns
```

```python
zomato['Country'].value_counts()
ind =zomato['Country'].value_counts().index
val =zomato['Country'].value_counts().values
sns.set(style="ticks", context="talk")
plt.style.use("dark_background")
plt.pie(val[:3],radius=3,labels=ind[:3])
plt.show()
```

**Observation :** Zomato maximum records or transaction are from indian after that USA and then UK.
The market place of zomato is india and india has big food market when compared to other places. 
The food delivery industry has more market place then other places  

```python
zomato.columns
```

```python
rating = zomato.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:"rating count"})
rating
```

```python
sns.set(style="ticks", context="talk")
plt.style.use("dark_background")
import matplotlib
matplotlib.rcParams['figure.figsize']=(20,9)
sns.barplot(x="Aggregate rating",y="rating count",data=rating)
```

```python
sns.countplot(x="Rating color",
data=rating,
palette=['blue','red','orange','yellow'
,'green','green'])
```

find the top 10 Cuisines?

```python
a=zomato['Cuisines'].value_counts().reset_index().rename(columns={'index':"Cuisines","Cuisines": "count Cuisines"})
Cuisines=pd.DataFrame(a).head(10)
sns.set(style="ticks", context="talk")
plt.style.use("dark_background")
plt.rcParams['figure.figsize']=(39,18)
sns.barplot(x="Cuisines",y="count Cuisines",data=Cuisines)
```
